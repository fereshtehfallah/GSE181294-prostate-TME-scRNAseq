Parse sample names and assign condition labels

```python

# Inspect all 63 CSV files and map to sample names + conditions
extract_dir <- "/mnt/shared-workspace/GSE181294/raw"
files <- list.files(extract_dir, pattern = "\\.csv\\.gz$", full.names = FALSE)

# Parse sample names from filenames: GSM*_<sample_name>.count.csv.gz
sample_info <- data.frame(
  filename = files,
  sample_name = sub("^GSM[0-9]+_(.+)\\.count\\.csv\\.gz$", "\\1", files),
  stringsAsFactors = FALSE
)

# Assign conditions based on sample name patterns
assign_condition <- function(name) {
  if (grepl("^HP", name)) return("normal")
  if (grepl("-N-LG$", name)) return("adjacent_LG")
  if (grepl("-N-HG$", name)) return("adjacent_HG")
  if (grepl("-T-LG$", name)) return("tumor_LG")
  if (grepl("-T-HG$", name)) return("tumor_HG")
  return(NA)
}

sample_info$condition <- sapply(sample_info$sample_name, assign_condition)

# Extract patient ID
extract_patient <- function(name) {
  if (grepl("^HP", name)) return(name)  # HP0-HP4 are their own patients
  m <- regmatches(name, regexpr("PCA[0-9]+", name))
  if (length(m) > 0) return(m) else return(NA)
}
sample_info$patient_id <- sapply(sample_info$sample_name, extract_patient)

# Summary
cat("Sample breakdown by condition:\n")
print(table(sample_info$condition))
cat("\nTotal samples:", nrow(sample_info), "\n")
cat("\nSample info:\n")
print(sample_info[order(sample_info$condition, sample_info$sample_name), c("sample_name","condition","patient_id")])
```

Load all 37 samples into Seurat with MAD-based QC filtering

```python

# Load all 37 samples into Seurat objects with QC metrics
# Strategy: read CSV, create Seurat, compute QC, filter, then merge

library(Seurat)
library(Matrix)
library(data.table)

extract_dir <- "/mnt/shared-workspace/GSE181294/raw"
sample_info_clean <- read.csv("/mnt/shared-workspace/GSE181294/sample_info.csv", stringsAsFactors = FALSE)
anno <- read.csv("/mnt/shared-workspace/GSE181294/cell_annotations.csv", stringsAsFactors = FALSE)

# Create lookup: cell_barcode -> cell_type
anno_lookup <- setNames(anno$cell_type, anno$cell_barcode)

seurat_list <- list()
qc_summary <- data.frame()

cat("Loading", nrow(sample_info_clean), "samples...\n")

for (i in seq_len(nrow(sample_info_clean))) {
  sname <- sample_info_clean$sample_name[i]
  fname <- file.path(extract_dir, sample_info_clean$filename[i])
  cond  <- sample_info_clean$condition[i]
  pid   <- sample_info_clean$patient_id[i]
  
  # Read CSV (genes x cells)
  mat <- fread(fname, header = TRUE, sep = ",", data.table = FALSE)
  rownames(mat) <- mat[, 1]
  mat <- mat[, -1]
  mat <- as.matrix(mat)
  storage.mode(mat) <- "integer"
  mat <- Matrix(mat, sparse = TRUE)
  
  # Create Seurat object
  so <- CreateSeuratObject(counts = mat, project = sname, min.cells = 3, min.features = 200)
  
  # Add metadata
  so$sample_id  <- sname
  so$condition  <- cond
  so$patient_id <- pid
  
  # Mitochondrial %
  so[["percent.mt"]] <- PercentageFeatureSet(so, pattern = "^MT-")
  
  # Attach published cell type annotation (where available)
  cb <- colnames(so)
  so$published_cell_type <- anno_lookup[cb]
  
  # QC before filtering
  n_before <- ncol(so)
  
  # MAD-based filtering (5 MADs per sample)
  nf <- so$nFeature_RNA
  nc <- so$nCount_RNA
  mt <- so$percent.mt
  
  mad_filter <- function(x, nmads = 5, type = "both") {
    med <- median(x, na.rm = TRUE)
    mad_val <- mad(x, na.rm = TRUE)
    if (type == "both") return(x >= (med - nmads * mad_val) & x <= (med + nmads * mad_val))
    if (type == "upper") return(x <= (med + nmads * mad_val))
    if (type == "lower") return(x >= (med - nmads * mad_val))
  }
  
  keep <- mad_filter(nf) & mad_filter(nc) & mad_filter(mt, type = "upper")
  so <- so[, keep]
  n_after <- ncol(so)
  
  seurat_list[[sname]] <- so
  
  qc_summary <- rbind(qc_summary, data.frame(
    sample_id  = sname,
    condition  = cond,
    patient_id = pid,
    cells_before = n_before,
    cells_after  = n_after,
    pct_retained = round(100 * n_after / n_before, 1),
    median_nFeature = round(median(so$nFeature_RNA)),
    median_nCount   = round(median(so$nCount_RNA)),
    median_pct_mt   = round(median(so$percent.mt), 2)
  ))
  
  if (i %% 5 == 0 || i == nrow(sample_info_clean)) {
    cat(sprintf("  [%d/%d] %s: %d -> %d cells (%.1f%% retained)\n",
                i, nrow(sample_info_clean), sname, n_before, n_after,
                100 * n_after / n_before))
  }
}

cat("\nAll samples loaded.\n")
cat("QC summary:\n")
print(qc_summary)
```

Load all 37 samples with fixed annotation lookup

```python

# The issue is that some cell barcodes in the Seurat object have NA names in the annotation lookup
# Fix: use match() instead of direct lookup to handle NAs properly

library(Seurat)
library(Matrix)
library(data.table)

extract_dir <- "/mnt/shared-workspace/GSE181294/raw"
sample_info_clean <- read.csv("/mnt/shared-workspace/GSE181294/sample_info.csv", stringsAsFactors = FALSE)
anno <- read.csv("/mnt/shared-workspace/GSE181294/cell_annotations.csv", stringsAsFactors = FALSE)

seurat_list <- list()
qc_summary <- data.frame()

cat("Loading", nrow(sample_info_clean), "samples...\n")

for (i in seq_len(nrow(sample_info_clean))) {
  sname <- sample_info_clean$sample_name[i]
  fname <- file.path(extract_dir, sample_info_clean$filename[i])
  cond  <- sample_info_clean$condition[i]
  pid   <- sample_info_clean$patient_id[i]
  
  # Read CSV (genes x cells)
  mat <- fread(fname, header = TRUE, sep = ",", data.table = FALSE)
  rownames(mat) <- mat[, 1]
  mat <- mat[, -1]
  mat <- as.matrix(mat)
  storage.mode(mat) <- "integer"
  mat <- Matrix(mat, sparse = TRUE)
  
  # Create Seurat object
  so <- CreateSeuratObject(counts = mat, project = sname, min.cells = 3, min.features = 200)
  
  # Add metadata
  so$sample_id  <- sname
  so$condition  <- cond
  so$patient_id <- pid
  
  # Mitochondrial %
  so[["percent.mt"]] <- PercentageFeatureSet(so, pattern = "^MT-")
  
  # Attach published cell type annotation using match (handles NAs safely)
  cb <- colnames(so)
  idx <- match(cb, anno$cell_barcode)
  so$published_cell_type <- anno$cell_type[idx]
  
  # QC before filtering
  n_before <- ncol(so)
  
  # MAD-based filtering (5 MADs per sample)
  nf <- so$nFeature_RNA
  nc <- so$nCount_RNA
  mt <- so$percent.mt
  
  mad_filter <- function(x, nmads = 5, type = "both") {
    med <- median(x, na.rm = TRUE)
    mad_val <- mad(x, na.rm = TRUE)
    if (type == "both") return(x >= (med - nmads * mad_val) & x <= (med + nmads * mad_val))
    if (type == "upper") return(x <= (med + nmads * mad_val))
    if (type == "lower") return(x >= (med - nmads * mad_val))
  }
  
  keep <- mad_filter(nf) & mad_filter(nc) & mad_filter(mt, type = "upper")
  so <- so[, keep]
  n_after <- ncol(so)
  
  seurat_list[[sname]] <- so
  
  qc_summary <- rbind(qc_summary, data.frame(
    sample_id  = sname,
    condition  = cond,
    patient_id = pid,
    cells_before = n_before,
    cells_after  = n_after,
    pct_retained = round(100 * n_after / n_before, 1),
    median_nFeature = round(median(so$nFeature_RNA)),
    median_nCount   = round(median(so$nCount_RNA)),
    median_pct_mt   = round(median(so$percent.mt), 2)
  ))
  
  if (i %% 5 == 0 || i == nrow(sample_info_clean)) {
    cat(sprintf("  [%d/%d] %s: %d -> %d cells (%.1f%% retained)\n",
                i, nrow(sample_info_clean), sname, n_before, n_after,
                100 * n_after / n_before))
  }
}

cat("\nAll samples loaded.\n")
```

Map clusters to published cell types via majority vote

```python

# ── Cluster-to-published-cell-type mapping ───────────────────────────────────
# Use published labels to annotate clusters (majority vote per cluster)
meta <- merged@meta.data

# Majority vote: for each cluster, find the most common published_cell_type
cluster_annotation <- meta %>%
  group_by(RNA_snn_res.0.6, published_cell_type) %>%
  summarise(n = n(), .groups = "drop") %>%
  group_by(RNA_snn_res.0.6) %>%
  slice_max(n, n = 1, with_ties = FALSE) %>%
  rename(cluster = RNA_snn_res.0.6,
         majority_cell_type = published_cell_type,
         n_cells = n)

cat("Cluster → Cell Type mapping (majority vote, res 0.6):\n")
print(cluster_annotation, n = 30)

# Also compute purity (% of majority type per cluster)
cluster_purity <- meta %>%
  group_by(RNA_snn_res.0.6) %>%
  summarise(
    total = n(),
    majority_type = names(sort(table(published_cell_type), decreasing=TRUE))[1],
    majority_n    = max(table(published_cell_type)),
    purity        = round(max(table(published_cell_type)) / n() * 100, 1),
    .groups = "drop"
  )
cat("\nCluster purity (% majority cell type):\n")
print(cluster_purity, n = 30)
cat("\nMean purity:", round(mean(cluster_purity$purity), 1), "%\n")
```

```python

# Read significant interactions to find top L-R pairs per condition
library(dplyr)

# Load all significant interactions
all_int <- do.call(rbind, lapply(
  c("normal","adjacent_LG","adjacent_HG","tumor_LG","tumor_HG"),
  function(cond) {
    f <- paste0("/mnt/results/06_cellchat/significant_interactions_", cond, ".csv")
    df <- read.csv(f)
    df$condition <- cond
    df
  }
))

cat("Total significant interactions across all conditions:", nrow(all_int), "\n")
cat("Columns:", paste(colnames(all_int), collapse=", "), "\n\n")

# Top L-R pairs by probability in tumor_HG
cat("=== Top 15 L-R pairs in tumor_HG (by prob) ===\n")
top_hg <- all_int %>% 
  filter(condition == "tumor_HG") %>%
  arrange(desc(prob)) %>%
  select(source, target, ligand, receptor, prob, pathway_name) %>%
  head(15)
print(top_hg)

# Interactions unique to tumor (not in normal)
cat("\n=== Interactions gained in tumor_HG (not in normal) ===\n")
normal_pairs <- all_int %>% filter(condition == "normal") %>%
  mutate(pair = paste(source, target, ligand, receptor, sep="|")) %>%
  pull(pair)
tumor_hg_unique <- all_int %>% filter(condition == "tumor_HG") %>%
  mutate(pair = paste(source, target, ligand, receptor, sep="|")) %>%
  filter(!pair %in% normal_pairs) %>%
  arrange(desc(prob)) %>%
  select(source, target, ligand, receptor, prob, pathway_name) %>%
  head(20)
cat("Unique to tumor_HG:", nrow(tumor_hg_unique), "interactions\n")
print(tumor_hg_unique)
```

Generate pathway activity heatmap across all conditions

```python

# Generate pathway presence heatmap across conditions
library(ggplot2)
library(dplyr)

diff_pw <- read.csv("/mnt/results/06_cellchat/comparative/differential_pathways.csv")
pw_summary <- read.csv("/mnt/results/06_cellchat/comparative/pathway_summary_comparative.csv")

conditions <- c("normal","adjacent_LG","adjacent_HG","tumor_LG","tumor_HG")
cond_labels <- c("Normal","Adjacent LG","Adjacent HG","Tumor LG","Tumor HG")

# Build long-format presence matrix with probability
pw_long <- pw_summary
pw_long$condition <- factor(pw_long$condition, levels = conditions,
                             labels = cond_labels)

# Order pathways: first by presence in normal, then by max probability
pw_order <- pw_summary %>%
  group_by(pathway) %>%
  summarise(max_prob = max(total_prob), n_cond = n()) %>%
  arrange(desc(n_cond), desc(max_prob)) %>%
  pull(pathway)

pw_long$pathway <- factor(pw_long$pathway, levels = rev(pw_order))

# Pathway presence heatmap (log-scaled probability)
p <- ggplot(pw_long, aes(x = condition, y = pathway, fill = log1p(total_prob * 1000))) +
  geom_tile(color = "white", linewidth = 0.3) +
  scale_fill_gradient(low = "#f7fbff", high = "#08306b",
                       name = "log(prob×1000)",
                       na.value = "grey95") +
  scale_x_discrete(labels = cond_labels) +
  labs(title = "Signaling pathway activity across disease stages",
       subtitle = "Color = log-scaled total communication probability | Grey = pathway absent",
       x = NULL, y = NULL) +
  theme_minimal(base_size = 9) +
  theme(axis.text.x = element_text(angle = 30, hjust = 1, size = 9, face = "bold"),
        axis.text.y = element_text(size = 7),
        plot.title = element_text(face = "bold", size = 11),
        plot.subtitle = element_text(size = 8, color = "grey40"),
        legend.position = "right",
        panel.grid = element_blank())

ggsave("/mnt/results/06_cellchat/comparative/pathway_activity_heatmap.png",
       p, width = 9, height = 14, dpi = 150)
cat("pathway_activity_heatmap.png saved\n")
cat("Pathways shown:", nlevels(pw_long$pathway), "\n")
```
